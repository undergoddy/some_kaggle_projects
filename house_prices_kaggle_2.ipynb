```python
import scipy.stats as st
```

```python
train_df_2 = train_2
```

```python
sns.distplot(train_df_2.SalePrice, fit=st.lognorm)
```

```python
st.probplot(train_df_2['SalePrice'], plot=plt)
```

```python
sns.distplot(train_df_2.GrLivArea, fit=st.norm)
```

```python
train_df_2.GrLivArea = np.log(train_df_2.GrLivArea)
```

```python
st.probplot(train_df_2.GrLivArea, plot=plt)
```

```python
sns.distplot(train_df_2.TotalBsmtSF, fit=st.norm)
```

```python
train_df_2['HasBsmt'] = pd.Series(len(train_df_2['TotalBsmtSF']), index=train_df_2.index)
train_df_2['HasBsmt'] = 0 
train_df_2.loc[train_df_2['TotalBsmtSF'] > 0, 'HasBsmt'] = 1
```

```python
train_df_2.loc[train_df_2['HasBsmt'] == 1, 'TotalBsmtSF'] = np.log(train_df_2['TotalBsmtSF'])
```

```python
sns.distplot(train_df_2[train_df_2['HasBsmt'] > 0]['TotalBsmtSF'], fit=st.norm)
```

```python
st.probplot(train_df_2[train_df_2['HasBsmt'] > 0]['TotalBsmtSF'], plot=plt)
```

```python
train_df_3 = train_df_2[['OverallQual', 'TotalBsmtSF', 'GrLivArea', 'SalePrice', 'HasBsmt']]
```

```python
train_df_3 = pd.get_dummies(train_df_3)
```

```python
train_df_3
```

```python
X_train = train_df_3.drop(columns=['SalePrice'])

y_train = train_df_3.SalePrice
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train, y_train)
```

```python
lin_reg.score(X_train, y_train)
```

```python
test_df_4 = test_df[['OverallQual', 'TotalBsmtSF', 'GrLivArea']]
```

```python
test_df_4.info()
```

```python
test_df_4.TotalBsmtSF.isna().value_counts()
```

```python
test_df_4.TotalBsmtSF.fillna(value=test_df_4.TotalBsmtSF.mean(), inplace=True)
```

```python
test_df_4['HasBsmt'] = pd.Series(len(test_df_4['TotalBsmtSF']), index=test_df_4.index)
test_df_4['HasBsmt'] = 0 
test_df_4.loc[test_df_4['TotalBsmtSF'] > 0, 'HasBsmt'] = 1
```

```python
test_df_4.loc[test_df_4['HasBsmt'] == 1, 'TotalBsmtSF'] = np.log(test_df_4['TotalBsmtSF'])
```

```python
test_df_4.GrLivArea = np.log(test_df_4.GrLivArea)
```

```python
test_df_4 = pd.get_dummies(test_df_4)
```

```python
test_df_4
```

```python
lin_reg_pred = pd.Series(lin_reg.predict(test_df_4))
```

```python
lin_reg_pred
```
